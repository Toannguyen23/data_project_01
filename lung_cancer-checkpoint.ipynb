```python
cancer_data=cancer_data.replace({'Level':{'Low': 1, 'Medium': 2, 'High': 3}})
```

```python
#huan luyen mo hình
import numpy as np
y = cancer_data['Level']
features = [col for col in cancer_data.columns if cancer_data[col].dtypes in ['float64', 'int64']]
X = cancer_data[features].drop('Level', axis = 1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X,y, train_size = .7)
model = LinearRegression()
model.fit(X_train, y_train)
preds = model.predict(X_val)
print(preds)
```

```python
print("R^2 có giá trị bằng: ", model.score(X_train, y_train))
```

```python
print("R^2 có giá trị trên mô hình test: ", model.score(X_val, y_val))
```
